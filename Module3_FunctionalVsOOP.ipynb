# OOP Programming
## Sort an Array of 0s, 1s, and 2s

```python
class Solution:
    def __init__ (self, nums):
        self.nums = nums

    def sort012(self):
        count0 = self.nums.count(0)
        count1 = self.nums.count(1)
        count2 = self.nums.count(2)

        self.nums[:] = [0] * count0 + [1] * count1 + [2] * count2
        return self.nums
```

# Functional Programming
## Binary Search

```python
arr = [0, 2, 1, 2, 0]
solution = Solution(arr)
solution.sort012()
```

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    compare = lambda x, y: x - y

    while left <= right:
        mid = (left+ right) // 2
        result = compare(arr[mid], target)

        if result == 0:
            return mid
        elif result < 0:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```

```python
arr = [1, 2, 3, 4, 5]
binary_search(arr, 4)
```
