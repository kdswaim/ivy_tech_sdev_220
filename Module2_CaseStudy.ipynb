```python
"""
File: Module2-CaseStudy.py
Author: Kristen Swaim
Created: 2026-01-21
Description: 
This program accepts student names and GPAs and determines
whether each student qualifies for the Dean's List (GPA equal or greater than 3.5) 
or Honor Roll (over 3.25 but less than 3.5) GPA
The program stops processing when the last name 'ZZZ' is entered.
"""

#student_name_last takes student's last name
student_name_last = input("Enter your last name: ")

while student_name_last != 'ZZZ':
    student_name_first = input("Enter your first name: ") #student_name_first stores student's first name
    student_gpa = float(input("Enter your GPA: ")) #student_GPA takes student's GPA

    #if student gpa greater or equal to 3.5, student is on dean's list
    #if student gpa greater or equal to 3.25 but less than 3.5, student is on honor roll list
    
    if student_gpa >= 3.5: 
        print(student_name_first, student_name_last, "has made the Dean's List.")
    elif student_gpa >= 3.25:
        print(student_name_first, student_name_last, "has made the Honor Roll.")

print()

student_name_last = input("Enter your last name: ")
```
