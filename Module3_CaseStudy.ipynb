```python
"""
Student Name: Kristen Swaim
File Name: Module3_CaseStudy.ipynb
Description:
This program uses classes and inheritance to store and display information
about a car. A Vehicle superclass holds the vehicle type, and an Automobile
subclass stores details such as year, make, model, number of doors, and roof
type. The program collects user input and prints the information in a
readable format.

Variables:
vehicle_type (str): The type of vehicle (car)
year (str): The year the car was made
make (str): The manufacturer of the car
model (str): The model of the car
doors (str): Number of doors (2 or 4)
roof (str): Type of roof (solid or sun roof)
"""

# Superclass
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type


# Subclass
class Automobile(Vehicle):
    def __init__(self, year, make, model, doors, roof):
        super().__init__("car")  # Store "car" in Vehicle
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof


# Main program
def main():
    print("Enter the following information about your car:")

    year = input("Year: ")
    make = input("Make: ")
    model = input("Model: ")
    doors = input("Number of doors (2 or 4): ")
    roof = input("Type of roof (solid or sun roof): ")

    car = Automobile(year, make, model, doors, roof)

    print("\nVehicle Information")
    print(f"Vehicle type: {car.vehicle_type}")
    print(f"Year: {car.year}")
    print(f"Make: {car.make}")
    print(f"Model: {car.model}")
    print(f"Number of doors: {car.doors}")
    print(f"Type of roof: {car.roof}")


if __name__ == "__main__":
    main()
```
