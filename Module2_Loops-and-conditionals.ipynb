4.1 - Number guessing exercise

```python
secret = 9
guess = 4

if guess == secret:
    print('just right')
elif guess > secret:
    print('too high')
elif guess < secret:
    print('too low')
```

6.2 - guess me comparison

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break

    number += 1
```
